## Requirements for ML Systems

- Reliability
    - System should perform the correct function at the desired level of performance (i.e. what happens when API fails? What happens when your request times out? What happens when your payload is not as expected?)

- Scalability
    - Scaling of model?
        - Horizontal or vertical?
    - One model or multiple?
        - Can the same model serve multiple use cases?
    - Model management scaling?
        - How to manage the many artifacts generated by your model pipeline?
        - How to monitor the multiple models?

- Maintainability
    - How to version the code, data and model artifacts you use?

- Adaptability
    - How well does your model adapt to new data and business requirements? 
    - How to ensure you adapt fast to such changes?

## ML Model Framing

- Framing is super important, because it directly affects maintainability of the model
- For example, imagine that I want to predict the next app that user is likely to open
    - Multiclass classification? Given all apps, predict probability that the next app is opened across all apps
        - This approach sucks. Because with every new app, you need to retrain your model
    - Instead, predict probability of app opening for every given app characteristic, then softmax across all apps

## Objective functions

- Usually, things start with straightforward application of loss function (e.g. cross entropy) $-\sum \text{label}_i \cdot \log{(\text{prob}_i)}$

- But then other objectives start coming in
    - e.g. maximise engagement while miniminising hate content. Then loss becomes $\text{loss} = \alpha \text{ cross entropy} + \beta \text{ quality loss}$

## Data alone is not enough
