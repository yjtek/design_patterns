# UML State Diagram

- State diagram is a technique to describe how your system behaves and responds to a series of events

- Each state has 3 pieces of information
    - state name
        - Meaningful name of the state
    - state variable
        - descriptive information about the state
    - activities
        - actions performed when in specific state
        - 3 types of activities
            - entry: actions that occur when you enter from another state
            - exit: actions that occur when you leave the state
            - do: actions that occur once/multiple times while object is in the state
