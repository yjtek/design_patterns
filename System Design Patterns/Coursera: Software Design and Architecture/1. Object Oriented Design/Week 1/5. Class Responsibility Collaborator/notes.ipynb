- In designing software, we want a way to plan out the software ahead of time, much like cue cards in organising a speech
    - Conceptual design helps identify the following
        - Components
        - Connections
        - Responsibilities

    - Technical design
        - Implementation details of the above

- To do this systematically, we use CRC (Class, Responsibility, Collaborator) cards

- Let's use a bank ATM machine as an example
    - You insert your card
    - Bank machine asks you to enter a pin to authenticate access
    - Once validated, you can choose to deposit, withdraw, or check balances
